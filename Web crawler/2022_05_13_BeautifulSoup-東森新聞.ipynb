### 取得上層標籤

```python
from bs4 import BeautifulSoup
import requests

for p in range(1,5,1):
  r1 = requests.get("https://news.ebc.net.tw/realtime",
            headers={
              "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36",
              "cookie":"",
              "referer":""},
              params={"page":p},
              data={}
  )
  p1 = BeautifulSoup(r1.text, "html.parser")
  ret = p1.find_all("span", {"class":"title"})
  for dd in ret:
    if dd.text != "":
      print(dd.parent.find("span",{"class":"small-gray-text"}).text, dd.text)
```
