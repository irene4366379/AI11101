```python
import requests
import csv

import io
import prettytable


r1 = requests.get("https://data.tycg.gov.tw/opendata/datalist/datasetMeta/download",
    headers={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36",
        "referer":"http://www.google.com.tw",
        "cookie":"PHPSESSID=l09qkttsnphr0nvooam98spj0t"},
    params={"id":"4163dad4-f1bb-4a78-a284-1c22a7ee9065",
        "rid":"13b24a77-c482-4266-af2b-9e2348d94ba2"},
    data={},)
r1.encoding = "utf-8"

f = io.StringIO(r1.text)

ret = list(csv.reader(f))
p1 = prettytable.PrettyTable(ret[0], encoding="utf-8")
for dd in ret[1:]:
  p1.add_row(dd)

print(p1)
```
